```python
import pandas as pd
housing = pd.read_csv("data/housing.csv")
housing.describe()
```

```python
import seaborn as sns
sns.scatterplot(x = "median_income", y = "median_house_value", data = housing)
```

```python
housing_preprocessed = housing[housing.median_house_value != 500001]
sns.scatterplot(x = "median_income", y = "median_house_value", data = housing_preprocessed)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

## Define a Model
### Create a linear regression object

```python
model = linear_model.LinearRegression()
```

## Train the Model

```python
model.fit(X=housing[["median_income"]], y = housing[["median_house_value"]])
```

## Evaluate the Model

```python
# print coefficients
print(f"Intercept: {model.intercept_}")
print(f"coefficients: {model.coef_}")
#y = mx(coefficient) + c (intercept)
```

```python
print(f"Mean squared error: {mean_squared_error(housing[['median_house_value']], model.predict(housing[['median_income']]))}")
print(f"Mean absolute error: {mean_absolute_error(housing[['median_house_value']], model.predict(housing[['median_income']]))}")
```
